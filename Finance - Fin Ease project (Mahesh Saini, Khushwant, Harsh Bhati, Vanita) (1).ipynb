```python
# ------------------ Import Libraries -------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# ---------------- Create Sample Dataset ----------------
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Income": [40000, 42000, 45000, 47000, 50000, 52000],
    "Expense": [20000, 22000, 25000, 27000, 30000, 32000]
}

df = pd.DataFrame(data)
```

```python
# ---------------  Add Savings Column -------------------
df["Savings"] = df["Income"] - df["Expense"]

print("Data:\n")
print(df)
```

```python
# ----------------- Basic Analysis ----------------------
print("\nTotal Income:", df["Income"].sum())
print("Total Expense:", df["Expense"].sum())
print("Total Savings:", df["Savings"].sum())
```

```python
# -------------- Visualization - Line Chart -------------
plt.figure()
plt.plot(df["Month"], df["Income"], marker='o', label="Income")
plt.plot(df["Month"], df["Expense"], marker='o', label="Expense")
plt.plot(df["Month"], df["Savings"], marker='o', label="Savings")
plt.title("Finance Overview")
plt.legend()
plt.show()
```

```python
# ------------- Visualization - Bar Chart ---------------
df.set_index("Month")[["Income", "Expense"]].plot(kind="bar")
plt.title("Income vs Expense")
plt.show()
```

```python
# ---------------- Seaborn Heatmap --------------------
df = pd.DataFrame({
    "Income": [20, 30, 40],
    "Expense": [15, 25, 35],
    "Savings": [5, 5, 5]
})

# Heatmap
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
# ----------- Machine Learning - Predict Future Expense ---------
X = np.array(range(len(df))).reshape(-1, 1)  # Months index
y = df["Expense"].values

model = LinearRegression()
model.fit(X, y)
```

```python
# ------------------ Predict next 3 months ------------------------
future_months = np.array(range(len(df), len(df)+3)).reshape(-1, 1)
predicted_expense = model.predict(future_months)

print("\n Predicted Future Expenses:", predicted_expense)
```

```python
# -------------- Plots actual vs predicted expenses --------------
df = pd.DataFrame({
    "Month": ["Apr", "May", "Jun"],
    "Expense": [2000, 2500, 3000]
})

y = df["Expense"]

# Plot
plt.figure()
plt.plot(df["Month"], y, marker='o', label="Actual Expense")

future_labels = ["Jul", "Aug", "Sep"]
predicted_expense = [3200, 3500, 3700]

plt.plot(future_labels, predicted_expense, marker='o', linestyle='dashed', label="Predicted Expense")

plt.legend()
plt.title("Expense Prediction")
plt.show()
```
